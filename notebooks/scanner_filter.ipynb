```python
# import pandas as pd

# path = r"C:\Users\joacosta\Dev\Python\ORF5\Data\export-9-20-2025, 4-04-27 PM\ObjectData.csv"

# # Read all columns first
# df = pd.read_csv(path, skiprows=5)
# print(f"Original shape: {df.shape}")

# # Drop columns where ALL values are NaN
# df_clean = df.dropna(axis=1, how='all')
# print(f"After removing all-NaN columns: {df_clean.shape}")
# print(f"Removed {df.shape[1] - df_clean.shape[1]} empty columns")

# # Filter for rows where Conditions column equals "NoRead" (note the space in column name)
# noread_rows = df_clean[df_clean[' Conditions'] == 'NoRead']

# # Extract Package Index values and format as 4-digit zero-padded strings
# package_indices = [f"{num:04d}" for num in noread_rows[' Package Index'].tolist()]

# print(f"\nFound {len(noread_rows)} rows with Conditions = 'NoRead'")
# print(f"Formatted Package Index values: {package_indices}")

# # Show unique values in Conditions column to verify
# print("\nUnique values in Conditions column:")
# print(df_clean[' Conditions'].value_counts())

import pandas as pd

path = r"C:\Users\joacosta\Dev\Python\ORF5\Data\export-9-20-2025, 4-04-27 PM\ObjectData.csv"

# Read all columns first
df = pd.read_csv(path, skiprows=5)
print(f"Original shape: {df.shape}")

# Drop columns where ALL values are NaN
df_clean = df.dropna(axis=1, how='all')
print(f"After removing all-NaN columns: {df_clean.shape}")
print(f"Removed {df.shape[1] - df_clean.shape[1]} empty columns")

# Filter for rows where Conditions column equals "NoRead" (note the space in column name)
noread_rows = df_clean[df_clean[' Conditions'] == 'NoRead']

# Extract Package Index values and format as 4-digit zero-padded strings
package_indices_noread = [f"{num:04d}" for num in noread_rows[' Package Index'].tolist()]

print(f"\nFound {len(noread_rows)} rows with Conditions = 'NoRead'")
print(f"NoRead Package Index values: {package_indices_noread}")

# Manual list of additional package indices
manual_list = ['1175', '1432', '1814', '1820', '1842', '0011', '0058', '0063', '0208', '0817', '1258', '1260', '1583', '1687', '1717', '1975', '1983', '1997', '0007', '0092', '0105', '0156', '0221', '0223', '0273', '0421', '0466', '0485']

# Merge both lists and remove duplicates
combined_list = list(set(package_indices_noread + manual_list))
combined_list.sort()  # Sort for easier reading

print(f"\nManual list count: {len(manual_list)}")
print(f"NoRead list count: {len(package_indices_noread)}")
print(f"Combined list count (after removing duplicates): {len(combined_list)}")

print(f"\nCombined Package Index values: {combined_list}")

# Show unique values in Conditions column to verify
print("\nUnique values in Conditions column:")
print(df_clean[' Conditions'].value_counts())
```
